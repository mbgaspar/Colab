```python
import numpy as np
import math
from scipy import special as sp
import matplotlib.pyplot as plt
```

```python
tstarr = 0 
tstop = 0.005
tstep = 1e-5
t = np.arange(tstarr,tstop,tstep)

s1 = 1
s2 = -1
fc = 1000

i = s1*np.cos(2*np.pi*fc*t)
q = s2*np.sin(2*np.pi*fc*t)

sinal_composto = i+q

plt.plot(t,sinal_composto)
plt.grid
```

```python
i_rx = sinal_composto*np.cos(2*np.pi*fc*t)
q_rx = sinal_composto*np.sin(2*np.pi*fc*t)
plt.figure()
l1, =plt.plot(t,i_rx,label="in-phase")
l2, =plt.plot(t,q_rx,label="quadratura")
plt.legend(handles=[l1,l2])

#como encontrar o angulo? 
#pode pegar o sinal s1 e s2 e fazer o arcotang entre eles
```

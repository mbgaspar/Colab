Questão 1:
Em um receptor de pulsos em banda base, sabe-se que a razão Eb/N0 é de 15 dB. Determine a BER para o caso de uma codificação NRZ unipolar e para o caso da codificação NRZ bipolar. 

```python
import numpy as np
from scipy import special as sp

def qfunc(arg):
    return 0.5-0.5*sp.erf(arg/np.sqrt(2))

EbNo_db=15
EbNo = np.power(10,(EbNo_db/10))
print(np.sqrt(EbNo))
nrz_u = qfunc(np.sqrt(EbNo))
print(format(nrz_u,"5.5e"))
nrz_p= qfunc(np.sqrt(2*EbNo))
print(format(nrz_p,"5.5e"))
```
